FGS coordinate systems 

```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm


%matplotlib inline
```

```python
image_file = '/user/kbrooks/itar/FGS/sim_imgs/acton_ims/GA1_m400.fits'

orig_image = fits.getdata(image_file)
header = fits.getheader(image_file, ext=0)
```

FGS image: In ITM/DMS/Sci frame

```python
origin = header['ORIGIN'].strip()

print("Origin: {}".format(origin))
```

# ITM/DMS Sci to FGS raw

### FGS Sci to FGS (1) RAW (raw-raw, real raw)

```python
image = orig_image.copy()

# FGS guider = 1; Swap axes
image = np.swapaxes(image, 0, 1)

fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].imshow(orig_image, cmap='viridis', norm=LogNorm(), extent=(0,2048,0,2048)) #origin in lower left
axes[0].set_title('FGS Science')
axes[0].arrow(50, 50, 250, 0, head_width=30, fc='white', ec='black', lw=4)
axes[0].arrow(50, 50, 0, 250, head_width=30, fc='white', ec='black', lw=4)
axes[0].annotate('X Sci', xy=(450, 50), xytext=(450, 50), color='black', weight='bold', size='xx-large')
axes[0].annotate('Y Sci', xy=(45, 450), xytext=(45, 450), color='black', weight='bold', size='xx-large')

axes[1].imshow(image, cmap='viridis', norm=LogNorm(), extent=(0,2048,0,2048))
axes[1].set_title('FGS Raw - G1')
axes[1].arrow(2000, 2000, -250, 0, head_width=30, fc='white', ec='black', lw=4)
axes[1].arrow(2000, 2000, 0, -250, head_width=30, fc='white', ec='black', lw=4)
axes[1].annotate('Y Raw', xy=(1350, 1950), xytext=(1350, 1950), color='black', weight='bold', size='xx-large')
axes[1].annotate('X Raw', xy=(1750, 1600), xytext=(1750, 1600), color='black', weight='bold', size='xx-large')
```

### FGS SCI to FGS (2) RAW 

```python
image = orig_image.copy()

# FGS guider = 2; Perform 90 degree rotation
image = np.rot90(image, k=1)

fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].imshow(orig_image, cmap='viridis', norm=LogNorm(), extent=(0,2048,0,2048)) #origin in lower left
axes[0].set_title('NIRCam Science')
axes[0].arrow(50, 50, 250, 0, head_width=30, fc='white', ec='black', lw=4)
axes[0].arrow(50, 50, 0, 250, head_width=30, fc='white', ec='black', lw=4)
axes[0].annotate('X Sci', xy=(450, 50), xytext=(450, 50), color='black', weight='bold', size='xx-large')
axes[0].annotate('Y Sci', xy=(45, 450), xytext=(45, 450), color='black', weight='bold', size='xx-large')

axes[1].imshow(image, cmap='viridis', norm=LogNorm(), extent=(0,2048,0,2048))
axes[1].set_title('FGS Raw - G2')
axes[1].arrow(2000, 50, -250, 0, head_width=30, fc='white', ec='black', lw=4)
axes[1].arrow(2000, 50, 0, 250, head_width=30, fc='white', ec='black', lw=4)
axes[1].annotate('Y Raw', xy=(1350, 50), xytext=(1350, 50), color='black', weight='bold', size='xx-large')
axes[1].annotate('X Raw', xy=(1750, 400), xytext=(1750, 400), color='black', weight='bold', size='xx-large')
```
